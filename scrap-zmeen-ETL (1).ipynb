```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
#https://www.zameen.com/Homes/Karachi_Gulistan_e_Jauhar-232-1.html
#https://www.zameen.com/Homes/Karachi_Gulistan_e_Jauhar-232-2.html
# Setup WebDriver
driver = webdriver.Chrome()
driver.get('https://www.zameen.com/Homes/Karachi_Gulistan_e_Jauhar-232-4.html')
time.sleep(5)  # Let page load

data = []

# Get all property cards
properties = driver.find_elements(By.CSS_SELECTOR, 'li.a37d52f0')

for index, prop in enumerate(properties):
    try:
        driver.execute_script("arguments[0].scrollIntoView();", prop)
        time.sleep(1)

        # Title
        try:
            title = prop.find_element(By.TAG_NAME, 'h2').text.strip()
        except:
            title = 'No Title'

        # Price
        try:
            price = prop.find_element(By.CSS_SELECTOR, 'span.dc381b54').text.strip()
        except:
            price = 'No Price'

        # Location
        try:
            location = prop.find_element(By.CSS_SELECTOR, 'div.db1aca2f').text.strip()
        except:
            location = 'No Location'

        # Area
        try:
            area_divs = prop.find_elements(By.CSS_SELECTOR, 'div.cb0c0514')
            if len(area_divs) > 1:
                area = area_divs[1].find_element(By.TAG_NAME, 'span').text.strip()
            else:
                area = 'No Area'
        except:
            area = 'No Area'

        # Bedrooms & Washrooms
        try:
            details = prop.find_elements(By.CSS_SELECTOR, 'span._6d9b9b83')
            bedroom = details[0].text.strip() if len(details) > 1 else '0'
            washroom = details[1].text.strip() if len(details) > 1 else '0'
        except:
            bedroom = '0'
            washroom = '0'

        # Date Added
        try:
            date_added = prop.find_element(By.CSS_SELECTOR, 'span.a018d4bd').text.strip()
        except:
            date_added = 'No Date'

        # Extract property link
        try:
            link_tag = prop.find_element(By.CSS_SELECTOR, 'a[aria-label="Listing link"]')
            property_link = link_tag.get_attribute('href')
            #property_link = 'https://www.zameen.com' + relative_link
        except:
            property_link = 'No Link'

        # Extract image link
        try:
            image = prop.find_element(By.CSS_SELECTOR, 'picture img').get_attribute('src')
        except:
            image = 'No Image'

        # Phone, Agent, Agency (click Call button)
        try:
            call_button = prop.find_element(By.CSS_SELECTOR, 'button._7c46df70')
            driver.execute_script("arguments[0].click();", call_button)

            wait = WebDriverWait(driver, 10)
            phone = wait.until(EC.presence_of_element_located((By.CLASS_NAME, '_1087688e'))).text.strip()

            try:
                agent = driver.find_element(By.CSS_SELECTOR, 'span[aria-label="Agent Name"]').text.strip()
            except:
                agent = 'No Agent'

            try:
                agency = driver.find_element(By.CSS_SELECTOR, 'span[aria-label="Agency Name"]').text.strip()
            except:
                agency = 'No Agency'

            time.sleep(1)

        except:
            phone = 'No Phone'
            agent = 'No Agent'
            agency = 'No Agency'

        # Save all data
        data.append({
            'Title': title,
            'Price': price,
            'Location': location,
            'Area': area,
            'Bedroom': bedroom,
            'Washroom': washroom,
            'Date_Added': date_added,
            'Image': image,
            'Phone': phone,
            'Agent': agent,
            'Agency': agency,
            'Image': image,          # Store image URL
            'Property_Link': property_link
        })

        time.sleep(1)  # Prevent blocking

    except Exception as e:
        print(f'❌ Error on property {index + 1}:', e)

# Create DataFrame
df = pd.DataFrame(data)
print(df)

#driver.quit()
```

```python
import pandas as pd
from datetime import datetime, timedelta
import re
```

```python
df.fillna('Unknown', inplace=True)
df.replace(['', 'None', 'No Price', 'No Area', 'No Title', 'No Location',
            'No date', 'No Image', 'No Phone', 'No Agent', 'No Agency', 'No Link'], 'Unknown', inplace=True)
print(df)
```

```python
# 2. Transform 'Price' column: Split into numeric price and unit
def extract_price_parts(p):
    match = re.match(r'([\d.]+)\s*(Lakh|Crore)', str(p))
    if match:
        number = float(match.group(1))
        unit = match.group(2)
        return number, unit
    return -1.0, '-1'  # default for missing or invalid entries

# Apply to DataFrame
df[['Price Value', 'Price Unit']] = df['Price'].apply(lambda x: pd.Series(extract_price_parts(x)))

# Convert numeric part to float (already done above but for safety)
df['Price Value'] = df['Price Value'].astype(float)

print(df)
```

```python
# 3. Transform 'Area': remove "Sq. Yd." and convert to number
df['Area'] = df['Area'].str.replace('Sq. Yd.', '', regex=False).str.strip()
df['Area'] = pd.to_numeric(df['Area'], errors='coerce').fillna(1).astype(int)
print(df)
# 4. Convert 'Date Added' to actual date
```

```python
def convert_date(text):
    text = str(text).replace('Added:', '').strip()
    now = datetime.now()

    if 'hour' in text:
        hours = int(re.findall(r'\d+', text)[0])
        return (now - timedelta(hours=hours)).date()
    elif 'minute' in text:
        minutes = int(re.findall(r'\d+', text)[0])
        return (now - timedelta(minutes=minutes)).date()
    elif 'day' in text:
        days = int(re.findall(r'\d+', text)[0])
        return (now - timedelta(days=days)).date()
    elif 'week' in text:
        weeks = int(re.findall(r'\d+', text)[0])
        return (now - timedelta(weeks=weeks)).date()
    else:
        return now.date()

df['Date_Added'] = df['Date_Added'].apply(convert_date)
print(df)
```

```python
# 5. Fix bedroom/washroom columns if needed
df['Bedroom'] = pd.to_numeric(df['Bedroom'], errors='coerce').fillna(0).astype(int)
df['Washroom'] = pd.to_numeric(df['Washroom'], errors='coerce').fillna(0).astype(int)

print(df)
```

```python
df.drop(columns=['Price'], inplace=True)
```

```python
# Optional: reorder columns
cols = ['Title', 'Price Value', 'Price Unit', 'Location', 'Area', 'Bedroom', 'Washroom',
        'Date_Added', 'Image', 'Phone', 'Agent', 'Agency', 'Property_Link']
df = df[cols]
print(df)
```

```python
df.columns = df.columns.str.replace(' ', '_')
print(df)
```

```python
# Export to CSV (ready for SQL or analytics)
df.to_csv('zameen_cleaned_4.csv', index=False)
```
